```python
%matplotlib inline
import numpy as np
```

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_66

images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/trainset/', verbose=True, 
                           max_images=10):
    
    i.crop_to_landmarks_proportion_inplace(1)
    i = i.rescale_landmarks_to_diagonal_range(100)
    labeller(i, 'PTS', ibug_face_66)
    if i.n_channels == 3:
        i = i.as_greyscale(mode='average')
    images.append(i)
```

# Deep Convolutional Kernel - Architecture 1

```python
from alaborticcv2015.deepconvkernel import DeepLogGaborConvKernelArch1

dcka1 = DeepLogGaborConvKernelArch1(patch_size=(7, 7), n_levels=5, padding='constant')
```

```python
dcka1.learn_network(images)
```

```python
from menpo.visualize import visualize_images

visualize_images(dcka1.filters[-1])
```

```python
dcka1.compute_network_response(images[0], level=1).view_widget()
```

```python
dcka1.compute_kernel_response(images[0]).view_widget()
```

```python
dcka1.learn_kernel(ext_shape=(1, 200, 200)).view_widget()
```
