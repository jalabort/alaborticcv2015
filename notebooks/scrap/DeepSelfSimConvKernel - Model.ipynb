```python
%matplotlib inline
import numpy as np
```

```python
import menpo.io as mio
from menpo.landmark import labeller, ibug_face_66

images = []
for i in mio.import_images('/data/PhD/DataBases/faces/lfpw/trainset/', verbose=True, 
                           max_images=10):
    
    i.crop_to_landmarks_proportion_inplace(1)
    i = i.rescale_landmarks_to_diagonal_range(100)
    labeller(i, 'PTS', ibug_face_66)
    if i.n_channels == 3:
        i = i.as_greyscale(mode='average')
    images.append(i)
```

# Deep Self Similarity Convolutional Kernel - Architecture 1

```python
from alaborticcv2015.deepconvkernel import DeepSelfSimConvKernelArch1

dcka1 = DeepSelfSimConvKernelArch1(patch_size=(7, 7), mean_centre=True)
```

```python
arch1_filters = dcka1.filters(images[1], level=3, group='ibug_face_66')
```

```python
from menpo.visualize import visualize_images

visualize_images(arch1_filters[-1])
```

```python
arch1_response = dcka1.compute_network_response(images[0], group='ibug_face_66')
```

```python
arch1_response.view_widget()
```

```python
k1_response = dcka1.compute_kernel_response(images[0], group='ibug_face_66')
```

```python
k1_response.view_widget()
```

```python
dcka1.learn_kernel(images[0], group='ibug_face_66', ext_shape=(1, 200, 200)).view_widget()
```
